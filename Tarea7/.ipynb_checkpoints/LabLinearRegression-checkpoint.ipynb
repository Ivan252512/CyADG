```python
import csv
import ast
from pandas import DataFrame
import pandas as pd
import numpy as np

def readFromFile(fileName):
    """This function reads a csv file and returns the content of the file as alist"""
    myfile = open(fileName, 'r')
    reader = csv.reader(myfile,delimiter='|',quoting=csv.QUOTE_NONE)
        
    myL = list(reader)
    
    return [ast.literal_eval(x) for x in myL[0]]
    
# Reads the ozone data for 2015 
# [0]-> ozonevalue 
# [1]-> id_estation 
# [2] -> Month
# [3] -> Year
file="MonthlyAverage.csv"
ozone = readFromFile(file)
```

```python
print(len(ozone))
```

```python
def problem1(ozono):
    datosOzono=[]
    for i in range(len(ozono)):
        datosOzono.append(ozono[i][0])
    return np.array(datosOzono)
print(problem1(ozone))
```

```python
from datetime import datetime 
def problem2(ozono):
    fechasOzono=[]
    for i in range(len(ozono)):
        fechasOzono.append(datetime(int(ozono[i][3]),int(ozono[i][2]),1))
    return np.array(fechasOzono)
print(problem2(ozone))
```

```python
import matplotlib.pyplot as plt

def problem3(x,y):
    fig=plt.figure(figsize=(30,8))
    plt.plot(x,y,color='r')
    plt.title('Monthly mean ozone in Mexico City 1986-2016')
    plt.xlabel('Dates')
    plt.ylabel('ppb (partes por billn)')
    plt.savefig('ozoneall.png',dpi=400) 
    return plt.show()
problem3(problem2(ozone),problem1(ozone))
```

```python
def problem4(anio,mes):
    x=anio+((mes-1)/12)
    return x
print(problem4(2015,1))
print(problem4(2015,6))
print(problem4(1994,9))
print(problem4(1999,4))
```

```python
def problem5(ozono):
    xLista=[]
    for i in range(len(ozono)):
        xLista.append(problem4(ozono[i][3],ozono[i][2]))
    return np.array(xLista)
print(problem5(ozone))
```

```python
from  numpy import dot
from  numpy.linalg import inv

def problem6(x,y):

    regresion=[]
    for i in range(len(ozone)):
        regresion.append([x[i],1])

    A = np.array(regresion) 
    invAtA = inv(np.dot(A.T,A)) 
    [nm, nb1] = dot( dot(invAtA, A.T), y) 
    print("Solution: m = ",nm, " b=",nb1)

    fig=plt.figure(figsize=(30,8))
    plt.plot(x,nm*x+nb1,color='b',linewidth=4)
    plt.plot(x,y,color='r')
    plt.title('Monthly mean ozone in Mexico City 1986-2016')
    plt.xlabel('Dates')
    plt.ylabel('ppb (partes por billn)')
    plt.savefig('ozoneall1.png',dpi=400)
    return plt.show()

problem6(problem5(ozone),problem1(ozone))
```

```python
def problem7(x,y):

    regresion=[]
    for i in range(len(ozone)):
        regresion.append([x[i]**2,x[i],1])
                         
    A = np.array(regresion) 
    invAtA = inv(np.dot(A.T,A)) 
    temp = dot(invAtA, A.T) 
    [na,nb2,nc] = dot(temp,y) 
    print("Solution: a = ",na, " b=",nb2, "c= ",nc)

    fig=plt.figure(figsize=(30,8))
    plt.plot(x,na*x*x+nb2*x+nc,color='b',linewidth=4)
    plt.plot(x,y,color='r')
    plt.title('Monthly mean ozone in Mexico City 1986-2016')
    plt.xlabel('Dates')
    plt.ylabel('ppb (partes por billn)')
    plt.savefig('ozoneall2.png',dpi=400)
    return plt.show()

problem7(problem5(ozone),problem1(ozone))
```

```python
def problem8(x,y):
    regresion1=[]
    for i in range(len(ozone)):
        regresion1.append([x[i],1])

    A = np.array(regresion1) 
    invAtA = inv(np.dot(A.T,A)) 
    [nm, nb1] = dot( dot(invAtA, A.T), y) 
    print("Solucin lineal: m = ",nm, " b=",nb1)
    x1=problem4(2018,5)
    print("Valores de ozono para mayo 2018=",nm*x1+nb1)
    x1=problem4(2021,5)
    print("Valores de ozono para mayo 2021=",nm*x1+nb1)
    
    
    
    regresion2=[]
    for i in range(len(ozone)):
        regresion2.append([x[i]**2,x[i],1])
                         
    A = np.array(regresion2) 
    invAtA = inv(np.dot(A.T,A)) 
    temp = dot(invAtA, A.T) 
    [na,nb2,nc] = dot(temp,y) 
    print("Solucin cuadrtica: a = ",na, " b=",nb2, "c= ",nc)
    x1=problem4(2018,5)
    print("Valores de ozono para mayo 2018=",na*x1*x1+nb2*x1+nc)
    x1=problem4(2021,5)
    print("Valores de ozono para mayo 2021=",na*x1*x1+nb2*x1+nc)

problem8(problem5(ozone),problem1(ozone))
```

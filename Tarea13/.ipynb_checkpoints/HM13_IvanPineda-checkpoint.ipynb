# Problem 5

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

#Convertimos sinclass en matriz para poder graficar aunque no es necesario.
X=sinclass.as_matrix()
#Obtenemos el kmeans para dos clusters.
kmeans = KMeans(n_clusters = 2, init = 'k-means++', random_state = 42) 
computekmeans=kmeans.fit(X)
#Esto es para graficar aunque no es necesario, fit predict calcula los centroides
y_kmeans = kmeans.fit_predict(X)

fig=plt.figure(figsize=(30,8))
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 50, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 50, c = 'blue', label = 'Cluster 2')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroides')
plt.title('Clusters')
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.legend()
plt.show()   
```

```python
#Las etiquetas de cada punto
kmeans = KMeans(n_clusters = 2, init = 'k-means++', random_state = 42).fit(X)
kmeans.labels_
```

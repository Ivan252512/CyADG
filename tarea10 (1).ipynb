```python
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.cm as cm

import numpy as np

from scipy import misc
from scipy import ndimage

from scipy import signal
```

```python
def Problem1(x):
    im1 = misc.imread(x)
    return im1

plt.imshow(Problem1("uno.jpeg"))
plt.show()
print(Problem1("uno.jpeg"))
```

```python
import pylab as plt
def Problem2(x):
    image_data = misc.imread(x).astype(np.float32)
    scaled_image_data = image_data / 255.
  
    red =  scaled_image_data[:,:,0]
    
    green =  scaled_image_data[:,:,1]
    
    blue =  scaled_image_data[:,:,2]
    
    print('Azul')
    print(blue)
    print('Verde')
    print(green)
    print('Rojo')
    print(red)
    
    plt.subplot(221)
    plt.imshow(red, cmap=plt.cm.Reds)
    plt.subplot(222)
    plt.imshow(green, cmap=plt.cm.Greens)
    plt.subplot(223)
    plt.imshow(blue, cmap=plt.cm.Blues)  
  
    return plt.show()

Problem2("uno.jpeg")
```

```python
def Problem3(x):

    I = misc.imread(x)
    
    kernel = np.array([[1, 1, 1],[1, 1, 1],[1, 1, 1]]) / 9.0
    
    r = signal.convolve2d(I[:,:,2],kernel, mode='same')
    
    plt.imshow(r,cmap="Greys_r")
    
    misc.imsave('greys.jpg', r)
Problem3("uno.jpeg")
```

```python
def Problem4(x):
    kersobx = np.array([[-1, 0,1],[-2, 0,2],[-1, 0, 1]])
    kersoby = np.array([[-1,-2,-1],[0, 0,0],[1,2,1]])

    im = misc.imread(x)

    
    t = np.array(im)

    # Transform image into grayscale
    greyI = np.zeros((len(t),len(t[0])))
    for row in range(len(t)):
        for col in range(len(t[0])):
            greyI[row,col] = t[row,col].mean()


    # Compute convolution between the two functions
    print( greyI.dtype)
    grad = signal.convolve2d(greyI, kersobx, boundary='symm', mode='valid')
    plt.subplot(2,2,1)
    plt.imshow(grad, cmap="Greys_r")

    grad1 = signal.convolve2d(greyI, kersoby, boundary='symm', mode='valid')
    plt.subplot(2,2,2)
    plt.imshow(grad1,  cmap="Greys_r")
p4('uno.jpeg')
```
